```python
def Search(searching, detales): # Paieska
    d1 = pd.DataFrame()
    search_tokens = []
    search_tokens.append(searching.split())
    relevant_word = relevant_words(search_tokens[0])
        
    k = 0
    match_index = set()
    for value in detales.Detalė.values:
        for i in value.split():
            if i.lower() in search_tokens[0]: # ziuriu ar yra tokios prekes atitankacias paieskai
                match_index.add(k)
            if i.lower() in relevant_word: # ziuriu ar atitinka panasiems zodziams
                match_index.add(k)
        k += 1

    k = 0
    for value in detales.daugiau_info.values:
        for i in value:
            if i.lower() in search_tokens[0]:
                match_index.add(k)
            if i.lower() in relevant_word:
                match_index.add(k)
        k += 1
            
    k = 0
    for value in detales.Info.values:
        for i in value.split():
            if i.lower() in search_tokens[0]:
                match_index.add(k)
            if i.lower() in relevant_word:
                match_index.add(k)
        k += 1
            
    for i in match_index: # sukuriu nauja dataframe, kur yra tik norimos eilutes
        d1 = d1.append(detales.iloc[i])
    
    return d1

def relevant_words(searching): # naudojant fasttext paketo lietuvisku zodziu vektoriu randami 30 panasiausi zodziai
    relevant_words = []
    for i in searching:
        relevant_words += ft.get_nearest_neighbors(i, k = 30)

    for i in range(len(relevant_words)):
        relevant_words[i] = relevant_words[i][1].lower()
    return relevant_words
```
